```python
cols = [
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI"
]

(data[cols] == 0).sum()
```

```python
import numpy as np

data[cols] = data[cols].replace(0, np.nan)
```

```python
data.isnull().sum()
```

```python
data.fillna(data.median(), inplace=True)
```

```python
data["Outcome"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    x="Outcome",
    data=data
)

plt.show()
```

```python
data.hist(
    figsize=(12,10)
)

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    data.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

Outlier Detection

```python
for col in data.columns[:-1]:

    plt.figure(figsize=(6,3))

    sns.boxplot(
        x=data[col]
    )

    plt.title(col)

    plt.show()
```
